```python
import matplotlib.pyplot as plt
```

```python
def cric_score():
    players = ["Virat Kohli", "MS Dhoni", "Rohit Sharma"]
    runs = [15000, 10500, 14000]
    
    plt.bar(players, runs)
    plt.title("Cricket Player Runs")
    plt.xlabel("Players")
    plt.ylabel("Runs")
    plt.show()

    return
```

```python
print(cric_score())
```
